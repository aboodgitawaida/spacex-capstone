```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import MarkerCluster

df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="FlightNumber", y="LaunchSite", hue="Class", palette="deep")
plt.title("Flight Number vs Launch Site by Success")
plt.xlabel("Flight Number")
plt.ylabel("Launch Site")
plt.show()

site_map = folium.Map(location=[28.5721, -80.648], zoom_start=5)
marker_cluster = MarkerCluster().add_to(site_map)

for index, row in df.iterrows():
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=f"Site: {row['LaunchSite']} - Success: {row['Class']}"
    ).add_to(marker_cluster)

site_map
```
