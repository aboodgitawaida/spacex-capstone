```python
import sqlite3
import pandas as pd

# Establish the connection and load data
con = sqlite3.connect("spacex_cache.db")
cur = con.cursor()

df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists="replace", index=False, method="multi")

print("--- Launch Sites Verification ---")
query_1 = "SELECT DISTINCT Launch_Site FROM SPACEXTBL;"
df_sites = pd.read_sql_query(query_1, con)
print(df_sites)

print("\n--- Successful Landings Count ---")
query_2 = "SELECT Mission_Outcome, COUNT(*) as Count FROM SPACEXTBL GROUP BY Mission_Outcome;"
df_outcome = pd.read_sql_query(query_2, con)
print(df_outcome)

con.close()
```
